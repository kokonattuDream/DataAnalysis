```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('../../data/currency_rate.xls',names=['Month','CR'])
df[0:5]
```

```python
newdf = df[0:10]
newdf
```

```python
dfpivot = newdf.pivot_table(values='CR',index=[newdf.index],columns=['Month'])
dfpivot
```

```python
dfpivot.fillna('',inplace=True)
dfpivot
```

```python
df1 = pd.DataFrame({'A': ['s1','s1','s1','s1','s1','s2','s2','s2','s2'],
                    'B': ['a','a','a','b','b','a','a','b','b'],
                    'C': ['cl1','cl0','cl0','cl1','cl1','cl0','cl1','cl1','cl0'],
                    'D': [1,4,6,5,12,32,43,11,4]})

df1
```

```python
df1.pivot_table(values='D',index=['A','B'],columns=['C'],aggfunc=np.sum)
```
