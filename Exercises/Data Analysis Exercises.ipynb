### 1) The common way to represent categorical variables is using the one-hot-encoding also known as dummy variables. The idea of dummy variables is to replace a categorical variable with  one or more new features that can have the values 0 and 1.
### Categories can be: workclass, flowers kinds, music kinds, car types, etc
### For example: You have a variable with 4 different kind of categories, that can be represented with numbers: 0, 1, 2, and 3 and this variable is represented by the following array:

```python
import numpy as np
y = np.array(([1], [0], [2], [3]), dtype=int)
```

### 1.1) Please use numpy library to make One Hot encoding matrix. 

Tip: Use the np.zeros to make the shape of the one hot matrix.

```python
N = len(y)
onehot = np.zeros((N, y.shape[0]))
print(onehot)
```

```python
list(enumerate(y))
```

```python
for id, val in enumerate(y):
    onehot[val,id] = 1.0
    
onehot    
```
