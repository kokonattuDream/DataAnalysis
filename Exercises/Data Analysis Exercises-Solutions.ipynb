### 1) The common way to represent categorical variables is using the one-hot-enconding also known as dummy variables. The idea of dummy variables is to replace a categorical variable with  one or more new features that can have the values 0 and 1.
### Categories can be: workclass, flowers kinds, music kinds, car types, etc
### For example: You have a variable with 4 different kind of categories, that can be represented with numbers: 0, 1, 2, and 3 and this variable is represented by the following array:

```python
import numpy as np
y = np.array(([1], [0], [2], [3]), dtype=int)
```

### Please use numpy library to make One Hot encoding matrix.

```python
N = len(y)
onehot = np.zeros((N,y.shape[0]))
print(onehot)
```

```python
print(list(enumerate(y)))
```

```python
for idx, val in enumerate(y):
    onehot[val,idx] = 1.0
    
onehot    
```

### 3-2) Select the columns satisfaction_level, last_evaluation, average_montly_hours, sales and salary. After that plot 1000 rows illustrating the relationships between variables for each kind of salary :

```python
hr_ = hr[['satisfaction_level','last_evaluation', 'average_montly_hours','sales','salary']].loc[:1000,:]
hr_
```

```python
sns.pairplot(hr_,palette='husl',hue='salary')
```

### 3-2) Make a boxplot graph illustrating the satisfaction level for each kind of salary and for each kind of sales department :

```python
g = sns.FacetGrid(hr_, col="sales", size=4, aspect=.5)
g = g.map(sns.boxplot, "salary", "satisfaction_level")
```
